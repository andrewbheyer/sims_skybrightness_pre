```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import lsst.sims.skybrightness_pre as sb
import healpy as hp
```

```python
sm = sb.SkyModelPre()
```

```python
hpid = 5000
plt.plot(sm.info['mjds'], sm.sb['r'][:,hpid], 'ko', alpha=.1)
moon_up = np.where(sm.info['moonAlts'] > 0)
plt.plot(sm.info['mjds'][moon_up], sm.sb['r'][:,hpid][moon_up], 'bo', alpha=.1)
sun_up = np.where(sm.info['sunAlts'] > np.radians(-13.))
plt.plot(sm.info['mjds'][sun_up], sm.sb['r'][:,hpid][sun_up], 'ro', alpha=.1)
```
